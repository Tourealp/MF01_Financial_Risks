```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python

data = pd.read_csv("Prediction_sheet.csv")
```

```python
data.head(10)
```

```python
data.boxplot()
```

```python
# X = data.iloc[:,0].values
# y = data.iloc[:,1].values

X = data[["m(g)"]]
y = data[["x(cm)"]]
```

```python
X,y
```

# Split datasets


```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.5,
                                                    random_state=3)
plt.scatter(X_test, y_test, color="b", label = 'Données vraies')
plt.scatter(X_train, y_train, color="r", label= 'Données prédites')
plt.legend()
plt.show()
```

# Regession model

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

# Predictive values

```python
y_pred = model.predict(X_test)
# print("y_test :", y_test)
# print("y_pred :\n", y_pred)

results = pd.DataFrame({
    "y_test": y_test.values.ravel(),
    "y_pred": y_pred.ravel()
})

print(results.head(10)) 
```

```python
plt.scatter(X_test, y_test, color="g", label = 'Données vraie')
plt.plot(X_train, model.predict(X_train), color='b', label="Droite linéaire prédictive")
plt.legend()
plt.show()
```
